```python
import pyaudio
import struct
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
plt.close('all')

#opens new window
%matplotlib tk

Format = pyaudio.paInt16
Channels = 1
Rate = 44100
Chunk = 1024*8

#pyaudio class isntance
p = pyaudio.PyAudio()

#stream objects to get data from the microphone
stream = p.open(
    format = Format,
    channels = Channels,
    rate = Rate,
    input = True,
    output = True,
    frames_per_buffer = Chunk
)

plt.title('SPECTROGRAM')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')

i=0
while True:
    #recieve binary data
    data = stream.read(Chunk, exception_on_overflow=False)
    
    #convert to int, make an array
    data_int = np.frombuffer(data, dtype = np.int16)
    
    #sets the variables using scipys spectrogram feature, f is frequency, t is time, sxx is strength
    f, t, Sxx = signal.spectrogram(data_int, Rate, nperseg=1024, nfft=1024, noverlap=0, mode='psd')

    #plots
    plt.ylim(-1,2000)
    plt.pcolormesh(t, f, Sxx)
    plt.title('SPECTROGRAM')
    plt.ylabel('Frequency [Hz]')
    plt.xlabel('Time [sec]')
    if i == 0:
        plt.clim(0,1)
        plt.colorbar()
        i+=1
    plt.show()
    

    plt.pause(0.00005)
    
```

```python
import pyaudio
import struct
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
plt.close('all')

#opens new window
%matplotlib tk

Format = pyaudio.paInt16
Channels = 1
Rate = 44100
Chunk = 1024*8

#pyaudio class isntance
p = pyaudio.PyAudio()

#stream objects to get data from the microphone
stream = p.open(
    format = Format,
    channels = Channels,
    rate = Rate,
    input = True,
    output = True,
    frames_per_buffer = Chunk
)

plt.title('SPECTROGRAM')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')

i=0
while True:
    #recieve binary data
    data = stream.read(Chunk, exception_on_overflow=False)
    
    #convert to int, make an array
    data_int = np.frombuffer(data, dtype = np.int16)
    
    #sets the variables using scipys spectrogram feature, f is frequency, t is time, sxx is strength
    plt.specgram(data_int,Fs=44100,pad_to=2048*2, NFFT=1048,noverlap=0)
    plt.ylim(0,2000)
    plt.show()

    plt.pause(0.00005)
    
```

```python
>>> from scipy import signal
>>> import numpy as np
>>> import matplotlib.pyplot as plt
>>> fs = 10e3
>>> N = 1e5
>>> amp = 2 * np.sqrt(2)
>>> noise_power = 0.01 * fs / 2
>>> time = np.arange(N) / float(fs)
>>> mod = 500*np.cos(2*np.pi*0.25*time)
>>> carrier = amp * np.sin(2*np.pi*3e3*time + mod)
>>> noise = np.random.normal(scale=np.sqrt(noise_power), size=time.shape)
>>> noise *= np.exp(-time/5)
>>> x = carrier + noise
>>> f, t, Sxx = signal.spectrogram(x, fs, nperseg = 1000)
>>> plt.pcolormesh(t, f, Sxx, shading='gouraud')
print(len(Sxx))
>>> plt.ylabel('Frequency [Hz]')
>>> plt.xlabel('Time [sec]')
>>> plt.show()
```

```python
#import the pyplot and wavfile modules 
import matplotlib.pyplot as plt
from scipy.io import wavfile
import numpy as np
from sklearn.decomposition import NMF

# Read the wav file (mono)
samplingFrequency, signalData = wavfile.read("gtr-nylon22.wav")

# Plot the signal read from wav file
plt.subplot(211)
plt.title('Spectrogram of a wav file with piano music')
plt.plot(signalData)
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.subplot(212)
plt.specgram(signalData,Fs=samplingFrequency)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.show()

#create matplot figure
fig, axs = plt.subplots(nrows=4, ncols=2, figsize=(15,8))
lines = []

#varible for plotting
x = np.linspace(0, len(signalData), 24)
x_fft = np.linspace(0, samplingFrequency, 129)

#create a line object pulling random data - can be random because it updates in a loop so y value doesn't matter
for i in range(4):
    lines.append(axs[i][0].plot(x, np.random.rand(24), '-', lw=2))
    lines.append(axs[i][1].semilogx(x_fft, np.random.rand(129), '-', lw=2))

#basic formatting for the plot
for i in range(4):
    axs[i][0].set_title('ACTIVATION (H)')
    axs[i][0].set_xlabel('Sample')
    axs[i][0].set_ylabel('Volume [dB]')
    axs[i][0].set_ylim(0,1.5)
    axs[i][0].set_xlim(0,len(signalData))

    axs[i][1].set_title('TEMPLATES (W)')
    axs[i][1].set_xlabel('Frequency [Hz]')
    axs[i][1].set_ylabel('Amplitude')
    axs[i][1].set_xlim(20, samplingFrequency / 2)

model = NMF(n_components=4, init='random', random_state=0)

signalData = np.abs(signalData)
print(signalData.shape)
W = model.fit_transform(signalData)
H = model.components_

print(len(W), len(W[0]))
print(len(H), len(H[0]))
      
i=0
j=0
while(j < 8):
    lines[j][0].set_ydata(H[i])
    j+=1
    lines[j][0].set_ydata(W[i])
    j+=1
    i+=1
    
fig.canvas.draw()

```
