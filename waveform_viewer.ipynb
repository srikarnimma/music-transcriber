```python
"""
pyaudio is used to capture the sound
numpy is used to convert the data from binary to ints
matplot is used to graph

np.frombuffer
"""

import pyaudio
import struct
import numpy as np
import matplotlib.pyplot as plt
import time

#opens new window
%matplotlib tk

#constants
Chunk = 1024                  #samples per frame
Format = pyaudio.paInt16        #bytes per sample
Channels = 1                    #one channel for mic
Rate = 44100                    #samples per second

#pyaudio class isntance
p = pyaudio.PyAudio()

#stream objects to get data from the microphone
stream = p.open(
    format = Format,
    channels = Channels,
    rate = Rate,
    input = True,
    output = True,
    frames_per_buffer = Chunk
)

#create matplot figure
fig, (ax, ax2) = plt.subplots(2, figsize=(15,8))

#variable for plotting
x = np.arange(0, 2*Chunk, 2)
x_cum = np.arange(0, 200*Chunk, 2)
y_cum = [0] * 100*Chunk

#create a line object - it doesn't matter what the values are rn
line, = ax.plot(x, np.random.rand(Chunk))
line_cum, = ax2.plot(x_cum, y_cum)

#basic formatting for the plot
ax.set_title('WAVEFORM VIEWER')
ax.set_xlabel('sample')
ax.set_ylabel('amplitude')
ax.set_ylim(-20000,20000)
ax.set_xlim(0,2*Chunk)

ax2.set_title('CUMULATIVE VIEWER')
ax2.set_xlabel('sample')
ax2.set_ylabel('amplitude')
ax2.set_ylim(-20000,20000)
ax2.set_xlim(0,200*Chunk)

plt.tight_layout()

start_time = time.time()
k=0
while True:
    data = stream.read(Chunk, exception_on_overflow=False)
    data_int = np.frombuffer(data, dtype = np.int16)
    y_cum = np.concatenate((y_cum[Chunk:], data_int), axis=None)
    
    line.set_ydata(data_int)
    line_cum.set_ydata(y_cum)
    
    
    fig.canvas.draw()
    fig.canvas.flush_events()
    
    
    
    
#     end_time = time.time()
#     ellapsed_time = end_time - start_time
#     print(k/ellapsed_time)
    
    
    k+=1


```

```python
"""
pyaudio is used to capture the sound
struct is used to convert the data from binary to ints
matplot is used to graph

~20fps
"""

import pyaudio
import struct
import numpy as np
import matplotlib.pyplot as plt
import time

#opens new window
%matplotlib tk

#constants
Chunk = 1024                  #samples per frame
Format = pyaudio.paInt16        #bytes per sample
Channels = 1                    #one channel for mic
Rate = 44100                    #samples per second

#pyaudio class isntance
p = pyaudio.PyAudio()

#stream objects to get data from the microphone
stream = p.open(
    format = Format,
    channels = Channels,
    rate = Rate,
    input = True,
    output = True,
    frames_per_buffer = Chunk
)

#create matplot figure
fig, (ax, ax2) = plt.subplots(2, figsize=(15,8))

#variable for plotting
x = np.arange(0, 2*Chunk, 2)
x_cum = np.arange(0, 200*Chunk, 2)
y_cum = [0] * 100*Chunk

#create a line object - it doesn't matter what the values are rn
line, = ax.plot(x, np.random.rand(Chunk))
line_cum, = ax2.plot(x_cum, y_cum)

#basic formatting for the plot
ax.set_title('WAVEFORM VIEWER')
ax.set_xlabel('sample')
ax.set_ylabel('volume')
ax.set_ylim(0,255)
ax.set_xlim(0,Chunk)

ax2.set_title('CUMULATIVE VIEWER')
ax2.set_xlabel('sample')
ax2.set_ylabel('amplitude')
ax2.set_ylim(0, 255)
ax2.set_xlim(0,200*Chunk)

start_time = time.time()
k=0
while True:
    data = stream.read(Chunk, exception_on_overflow=False)
    data_int = np.array(struct.unpack(str(2*Chunk) + 'B', data), dtype = 'b')[::2] +128
    y_cum = np.concatenate((y_cum[Chunk:], data_int), axis=None)
    
    line.set_ydata(data_int)
    line_cum.set_ydata(y_cum)
    
    fig.canvas.draw()
    fig.canvas.flush_events()
    
    
#     end_time = time.time()
#     ellapsed_time = end_time - start_time
#     print(k/ellapsed_time)
    
    
    k+=1


```
